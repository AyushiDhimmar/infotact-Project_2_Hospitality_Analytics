# Feature Engineering

This notebook focuses on creating new features and transforming existing variables to improve analytical quality and predictive modeling performance.

The objective of feature engineering is to:
- create meaningful business features
- improve model interpretability
- prepare data for machine learning workflows
- enhance hospitality analytics insights

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/cleaned/cleaned_hotel_bookings.csv")
```

## Creating Total Stay Feature

This feature combines weekday and weekend stay durations to calculate the total number of nights spent by a guest during a hotel booking.

```python
df['total_stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
```

## Creating Total Guests Feature

This feature calculates the total number of guests involved in a booking by combining adults, children, and babies.

```python
df['total_guests'] = df['adults'] + df['children'] + df['babies']
```

```python
df[['adults', 'children', 'babies', 'total_guests']].head()
```

## Creating Booking Season Feature

This feature categorizes bookings into seasons based on arrival month to support seasonal demand analysis and pricing insights.

```python
def get_season(month):
    if month in ['December', 'January', 'February']:
        return 'Winter'
    elif month in ['March', 'April', 'May']:
        return 'Spring'
    elif month in ['June', 'July', 'August']:
        return 'Summer'
    else:
        return 'Autumn'

df['booking_season'] = df['arrival_date_month'].apply(get_season)
```

```python
df[['arrival_date_month', 'booking_season']].head()
```

## Creating Lead Time Category Feature

This feature categorizes customers based on booking lead time to better understand early planners and last-minute booking behavior.

```python
def categorize_lead_time(days):
    if days <= 30:
        return 'Short Lead Time'
    elif days <= 90:
        return 'Medium Lead Time'
    else:
        return 'Long Lead Time'

df['lead_time_category'] = df['lead_time'].apply(categorize_lead_time)
```

```python
df[['lead_time', 'lead_time_category']].head()
```

## Feature Engineering Summary

Several new features were created to enhance the dataset for analytical and predictive modeling purposes.

Created features include:
- `total_stay`
- `total_guests`
- `booking_season`
- `lead_time_category`

These engineered variables provide additional business context related to customer behavior, travel seasonality, and booking patterns.

## Export Feature Engineered Dataset

The updated dataset containing engineered features is exported for further predictive modeling and dashboard development.
